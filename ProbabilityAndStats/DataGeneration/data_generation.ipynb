```python
# Useful imports
import matplotlib.pyplot as plt
import pandas as pd
```

# Data Generation

- Useful for testing out our models
- Quick!

Note that we want our data to be realistic (random)

## Code some examples

```python
# For plotting our examples
colors = {0:'red', 1:'blue', 2:'green', 3:'yellow'}

def plot_data(data_X, data_Y, colors=colors):
    df = pd.DataFrame(dict(x=data_X[:,0], y=data_X[:,1], label=data_Y))
    grouping = df.groupby('label')
    fig, ax = plt.subplots()
    for key, group in grouping:
        group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
    plt.show()
```

```python
from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=100, centers=4, n_features=2)
```

```python
plot_data(X, y)
```

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, noise=0.1)
```

```python
plot_data(X, y)
```

```python
from sklearn.datasets import make_circles
X, y = make_circles(n_samples=500, noise=0.05)
```

```python
plot_data(X, y)
```

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=1000, n_features=1, noise=15)

a = 0.1
plt.scatter(X,y,alpha=a)
plt.show()
```

### Modifying the Regression

```python
y2 = y**2
y3 = y**3

# plot regression dataset
a = 0.3

plt.scatter(X,y2,alpha=a)
plt.show()
plt.scatter(X,y3,alpha=a)
plt.show()
```
